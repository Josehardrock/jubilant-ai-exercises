**Tuplas**

```python
# Create an empty tuple with the built-in python declaration
empty_tuple = tuple

# Create a  tuple with different data types (more than one)
dragons_numbers_tuple = ("Drogon", "Vermax", "SilverFire", "Vermitor", "Syrax", 1, 2, 3, 4, 5)

# Print the second value of the new tuple
print(f"The second value of the tuple {dragons_numbers_tuple[1]}")

# Assing those values to multiple variables and print them

third_dragon = dragons_numbers_tuple[2]
print(f"The third dragon is {third_dragon}")

only_dragons = dragons_numbers_tuple[0:5]
print(f"The dragons are {only_dragons}")

only_numbers = dragons_numbers_tuple[5:]
print(f"The numbers are {only_numbers}")

# Add a value to a tuple
dragons_numbers_tuple = ("Balerion",) + dragons_numbers_tuple

# Print the second value of the tuple
print(f"The second value of the tuple is {dragons_numbers_tuple[1]}")
print(f"The new tupple is {dragons_numbers_tuple}")

```
