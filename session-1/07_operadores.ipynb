```python
# TODO write a program that ask for a 24 hour date and conver it into 12 hour date.
# Ex. Input -> 17 Output -> 5

def to_twelve_hours_format (time:str) -> str:
    if time < 0 or time > 24:
        return "Invalid time"
    original_time = time
    expected_time = 0

    am_pm = "am" if original_time < 12  else  "pm"
    
    if original_time > 12  and original_time < 24 : 
            expected_time = original_time - 12 
    elif original_time == 24 : 
        expected_time =  0    
    else : 
        expected_time = original_time

    twelve_time_format = f"Input time :  {original_time}, Output time --> {expected_time} {am_pm}"
    return twelve_time_format

# Add parenthesis to the expression to match the following output ---> 7:
print(to_twelve_hours_format(19))
print(to_twelve_hours_format(12))
print(to_twelve_hours_format(24))
print(to_twelve_hours_format(0))    
print(to_twelve_hours_format(13))    
print(to_twelve_hours_format(5))

```
