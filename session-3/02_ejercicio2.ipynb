```python
# Ejercicio 2
# •	Mediante la función map() y una función anónima convierte la siguiente lista de temperaturas en grados centigrados a grados 
#  kelvin. Siendo que la forma de pasar de grados centigrados a grados kelvin es sumar a la cifra 273.15
# 
# temperatures = [-23, 0, 34, 22, -30, 23]
# tem_kelvi = [250.15, 273.15, 307.15, 295.15, 243.15, 296.15]
# 
# •	Repite el procedimiento con list comprehension

temperatures = [-23, 0, 34, 22, -30, 23]

# Anonym function
toKelvin = lambda t : t + 273.15
temperatures_kelvin = []

for index, item in enumerate(temperatures) :
    temperatures_kelvin.append(round(toKelvin(item), 2))

print(f"Using anonymus, Input temperatures {temperatures} -> Temperatures in kelvin {temperatures_kelvin}")
```

```python
# Using map()

temperatures_kelvin = list(map(lambda t : round(t + 273.15, 2), temperatures))
print(f"Using map(), Input temperatures {temperatures} -> Temperatures in kelvin {temperatures_kelvin}")
```

```python


# List comprehension
temperatures_kelvin = [round(toKelvin(item), 2) for item in temperatures]

print(f"Using list comprehension(), Input temperatures {temperatures} -> Temperatures in kelvin {temperatures_kelvin}")
```
