```python
# Ejercicio 5
# En este ejercicio os voy a obligar a mirar parte de la documentación oficial de algunas de las librerías mencionadas.
# •	Numpy --> Escribe un programa que importe numpy y enseñe la versión y su configuración. Posteriormente crea un array con np.array() 
# y comprueba si alguno de los elementos tiene un 0 con la función np.all(). Ahora crea un array con 6 valores aleatorios 
# entre el 0 y el 1.

import numpy as np

print(f"Numpy Version {np.version.version}")
print(f"Numpy Config {np.show_config()}")

num_array = np.array([1,2,3,4,5,6])

print(f"Array: {num_array}")

contains_zero = np.all(num_array == 0)
print(f"Contains Zero: {contains_zero}")
num_random_array = np.array(np.random.randint(0,1))
print(f"Random Array: {num_random_array}")
```

```python

# •	Pandas --> ¿Qué son los DataFrames?. 
# Son estructuras de datos de tipo tabla, como las tablas de SQL, contienen dos dimensiones e incluyen filas y columnas.


# Escribe un programa que importe pandas y numpy. Primero declara un array de 4x3, siendo 
# # el primer valor nombres de ciudades, el segundo su población y el tercero el precio por metro cuadrado del terreno (pueden ser # datos inventados). 
# Ahora declara una lista con tres valores para las tres columnas. Crea un dataframe con los valores y las 
# # columnas que acabas de crear.

import pandas as pd
import numpy as np

cities_array = np.array([["Madrid", "Barcelona", "Valencia", "Sevilla"], 
                                    [3000000, 2000000, 1000000, 800000], 
                                    [2000, 1500, 1000, 800]])
cities_columns = ["City", "Population", "Pricem2"]

cities_dataframe = pd.DataFrame(cities_array, cities_columns)

print(f"Cities DataFrame: {cities_dataframe}")
```
