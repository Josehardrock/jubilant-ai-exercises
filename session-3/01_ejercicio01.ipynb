```python
# Ejercicio 1
# •	¿Para que sirven los argumentos de longitud variable? 
# Dan versatilidad a la función, permitiendo que esta pueda recibir un número variable de argumentos, sin necesidad de definirlos de antemano.
# También permite escribir una sola función sin tener que sobrecargarla cada vez que se necesite un número diferente de argumentos.

# Escribe una función que acepte un número indeterminado de argumentos. 
#  Estos argumentos serán las capas de un módelo, para ello iremos iterando por todos los argumentos, comprobando si cada capa es 
#  del tipo correcto (tienen que ser strings) e imprimiendo su valor.

def checkModel(*args):
    for arg in args:
        if isinstance(arg, str):
            print(f"Argument is valid:  {arg}")
        else:
            print(f"Invalid Argument: {arg} is not a string")

checkModel( 1, 2, 3,  4, 5, 6, "layerx", 7)
checkModel("layer1", "layer2", "layer", 4, 5, 6)
checkModel("layer1", "layer2", "layer3", "layer4", "layer5")
```
