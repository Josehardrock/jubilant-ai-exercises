```python
# ________________________________________
# Ejercicio 3
# •	Escribe una función en Python que invierta una cadena de texto. Ej "Hola mundo" quedaría "odnum aloH". 
# La función debe # comprobar que el parámetro de entrada sea una cadena y si no devolver un error.

def reverse_string(text:str) -> str :
    if not isinstance(text, str):
        raise TypeError("Invalid input")
    return text[::-1]

print(reverse_string("Hola mundo")) # odnum aloH
print(reverse_string("anilina")) # anilina
```
