```python
# Ejercicio 4
# •	Define una clase llamada RedNeuronal con las siguientes características:
# o	Debe tener un constructor sin parámetros de entrada
# o	El constructor debe definir un atributo llamado capas, que va a ser una lista vacía
# o	Va a existir un método llamado agregar_capa, que aceptará un parámetro y añadirá la capa al atributo capa
# o	Tendremos un método llamado forward que nos permita iterar por todas las capas, y cada vez que iteremos, imprimiremos la # capa por pantalla.
# •	Ahora instancia esta clase, añade dos capas (pueden ser cadenas que representen capas) y ejecuta el paso forward


class NeuralNetwork:
    def __init__(self):
        self.layers = []

    def add_layer(self, layer):
        self.layers.append(layer)

    def forward(self):
        for layer in self.layers:
            print(layer)

network = NeuralNetwork()
network.add_layer(["texto1", "texto2"])
network.add_layer('Layer 2')
network.add_layer(["Daemon", "Vermax", "Aemond" , "Sunfyre"])
network.forward()

```
