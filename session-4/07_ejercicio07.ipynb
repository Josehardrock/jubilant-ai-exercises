Utiliza Matplotlib para crear un histograma de la variable 'A'.

```python


import pandas as pd 
import matplotlib.pyplot as plt

input_data = {'A': [1, 2, 2, 3, 3, 3, 4, 4, 5]}
dframe = pd.DataFrame(input_data)

plt.hist(dframe, bins=50)

# TÃ­tulo y etiquetas
plt.title('Histograma')
plt.xlabel('Variable A ')
plt.ylabel('Frecuencia')

# Mostrar el grÃ¡fico
plt.show()
```
