```python
# Ejercicio 5
# Escribe un fragmento de código que intente dividir dos números y 
# maneje la excepción que se produce cuando el denominador es cero.

def divide(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return "A value can't be divided by zero"

def test_divide(x, y):
    print(f"Testing values {x} and {y}")
    print(f"Operation result :  {divide(x,y)}")


test_divide(10, 2)
test_divide(10, 0)

```
