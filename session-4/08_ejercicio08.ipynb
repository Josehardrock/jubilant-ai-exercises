•	Utiliza la biblioteca Transformers de Hugging Face para cargar un pipeline de análisis de sentimientos. Luego, analiza el sentimiento de la siguiente frase: "I love learning about machine learning with Python!".


```python


from transformers import pipeline

feelings_pipeline = pipeline('sentiment-analysis')

result = feelings_pipeline("I love learning about machine learning with Python!")

print(f"The sentiment is {result}")
```
