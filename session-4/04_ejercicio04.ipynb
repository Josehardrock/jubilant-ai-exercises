Ejercicio 4
•	Investiga y escribe sobre las diferencias clave entre PyTorch y TensorFlow, centrándote en aspectos como la definición de modelos, la ejecución de operaciones y la gestión de la memoria.

# Pytorch: 

### Gestion de Modelos: 

* Los modelos creados necesitan implementar dos metodos heredados de la clase nn.Module, el __init__ para definir las propiedades del model, y el "forward" para definir el flujo de datos del modelo. 

### Ejecución de Operaciones:

* Las operaciones son definidas en mode "Eager" que se traduce como definir -> ejecutar. Cada operación se ejecuta despues de definirla. (Es más sencillo encontrar los errores del código, depurar).

### Gestion de Memoria: 

* La propriedad básica de Pytorch es el objeto "Tensor", el framework maneja automaticamente la asignación y liberación de memoria.
* Uso de las GPU, pytorch utiliza las GPU disponible y puede compartir (distribuir) las operaciones en base a las GPU disponibles.

#### Conclusiones

* Se puede entender más facil de entender para un programador, permite experimentar de forma más sencilla, sin tener que indagar demasiado en las capaz del framework. 

# TensorFlow 

### Gestion de Modelos: 

* La creación de modelos se hace de una forma más estática, se crea una variable y se le pasan parametros para su definición, no requiere definición del method __init__ o forward, estos son manejados directamente via el framework. 

* API de Keras: 

### Ejecución de Operaciones:

* Funcionaba en modo "gráfico" lo cual implicaba generar un grafo, los cuales necesitaban una sesión. Cada sesión asignaba los recursos necesarios y ejecutaba las operaciones. 
 
* Mode Eager, TensorFlow posee un mode Eager el cual ejecuta las operaciones de manera inmediata en modo definir -> ejecutar. 

### Gestion de Memoria: 

* El framework maneja automaticamente la asignación y liberación de memoria.
* Utiliza XLA (Accelerated Linear Algebra) para compilar subgrafos en código de bajo nivel, lo cual permite optimizar el uso del hardware. 

# Conclusiones

* Se puede entender más facil de entender para un programador, permite experimentar de forma más sencilla, sin tener que indagar demasiado en las capaz del framework.
