```python
#Ejercicio 2
# •	Dado un vector en NumPy, realiza las siguientes tareas:
# 1.	Filtra los elementos mayores a 5 usando un bucle for.
# 2.	Realiza la misma operación utilizando list comprehension.
# 3.	Realiza la operación utilizando una función anónima.
#       Puedes utilizar este de ejemplo:
#       vector = np.array([1, 6, 3, 8, 4, 7])


import numpy as np
planet_position = np.array([1, 6, 3, 8, 4, 7])

beyond_jupyter = planet_position[planet_position > 5]
print(f"Using a simple filtering operation {beyond_jupyter}")

# Using a lambda function
beyond_jupyter = list(filter(lambda x: x > 5, planet_position))
print(f"Using a lambda function {beyond_jupyter}")

# Using a list comprehension
beyond_jupyter = [x for x in planet_position if x > 5]
print(f"Using a list comprehension {beyond_jupyter}")

```
