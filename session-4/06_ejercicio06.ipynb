## API Secuencial

* Es la forma más sencilla de crear modelos lineales, como su nombre lo indica se trata de una secuencia en la cual cada capa va a apilarse encima de la anterior. Cubre las necesidades básicas de las redez neuronales. 

( Es el metodo más sencillo para aprender a utilizar las API de Keras, al menos el punto de entrada ).


## API Funcional

* Cada capa se considera como una función con un tipo de entrada y de salida. Fue creada para topologías mas avanzadas. Permite crear modelos mas complejos (no solo lineales) con multiples entradas y salidas. Permite creer capas que se pueden reutilizar en diferentes partes del modelo. 

## API Basada en modelos

* Es una API basada en modelos "clases" que permite crear una clase que representa el modelo de la red neuronal. Se pueden crear capaz de cualquier tipo y se puede establecer conexiones entre las capas.

Es la API mas personalizable de Keras, se puede personalizar el modelo, y se tiene mucho controlo sobre el flujo de datos. Se puede definir su propia función de perdida si es necesario, y también se puede escoger o crear el optimizador deseado.

```python
# API Secuencial

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Crear un modelo secuencial
model = Sequential()

# Agregar capas   

model.add(Dense(64, activation='relu', input_dim=100))
model.add(Dense(10, activation='softmax'))
```

```python
# API Basada en Modelos

import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model

class MyModel(Model):
  def __init__(self):
    super(MyModel, self).__init__()
    self.dense1 = Dense(64, activation='relu')
    self.dense2 = Dense(10, activation='softmax')

  def call(self, inputs):
    x = self.dense1(inputs)
    return self.dense2(x)

# Crear una instancia del modelo
model = MyModel()

# Compilar el modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# API funcional de Keras

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Capa de entrada
inputs = Input(shape=(784,))

# Primera capa oculta
x = Dense(64, activation='relu')(inputs)

# Segunda capa oculta
x = Dense(64, activation='relu')(x)

# Capa de salida
predictions = Dense(10, activation='softmax')(x)

# Crear el modelo
model = Model(inputs=inputs, outputs=predictions)
```
