```python
# Programa una función que imprima todos los números de un rango que se le pase por parámetros que sean divisibles por 7. 
# Si el rango superior es menor que el rango inferior o si los argumentos no son números imprime un mensaje de error.
# imprime_rango(1, 20) -> ✅
# imprime_rango(20, 10) -> ❌
# imprime_rango(“hola”, 2) -> ❌

def printRangeDivisibleBySeven(lower:int, higher:int):
    print("\n ")
    if type(lower) != int or type(higher) != int:
        print("Invalid Range❌")
        return
    
    elif higher < lower:
        print("Invalid Range❌")
        return
     
    print(f"Valid range {lower}, {higher}✅") 
    for item in range(lower, higher):
        item_modulus = item % 7
        if item_modulus == 0 : print(f"Item with value {item} is divisible by 7 ✅")

range_1 = 1,20
range_2 = 20, 10
range_3 = "hola", 55


print(f"Test cases {range_1} is, result is {printRangeDivisibleBySeven(range_1[0], range_1[1])}")
print(f"Test cases {range_2} is, result is {printRangeDivisibleBySeven(range_2[0], range_2[1])}")
print(f"Test cases {range_3} is, result is {printRangeDivisibleBySeven(range_3[0], range_3[1])}")   

```
