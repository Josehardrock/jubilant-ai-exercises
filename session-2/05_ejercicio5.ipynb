```python
#	Escribe un script que pida por pantalla (con la función input()) una serie de números 
#  separados por comas y los convierta a una Lista y una Tupla:

# Input: 3, 6, 9, 14, 22
# Output: List -> [3, 6, 9, 14, 22] / Tuple -> (3, 6, 9, 14, 22)

def input(*args):

    list_numbers = list(args)
    tuple_numbers = tuple(list_numbers)
    print(f"List -> {list_numbers} / Tuple -> {tuple_numbers}")


input(1,2,6,7,8,9,10)


```
