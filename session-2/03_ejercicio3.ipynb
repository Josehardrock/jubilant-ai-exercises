```python
# Crea un diccionario con dos claves. La primera va a ser el nombre de ciudades y la segunda la población de estas mismas. 
# En la clave de las ciudades crearemos una lista con 3 o 4 valores correspondiente a nombres de ciudades. 
# En la clave de la población pondremos en orden el número de población de estas ciudades. Tiene que quedar algo como esto:


dataframe = { "ciudades": ["madrid", "segovia", "sevilla", "barcelona"], "poblacion": [40000, 34000, 23000, 50000] }

#Ahora convierte este diccionario en otro que tenga como clave el nombre de las ciudades y como valor su población (pista, podéis usar enumerate para conseguir el índice)
# {'madrid': 40000, 'segovia': 34000, 'sevilla': 23000}


cities_dictionnary = { dataframe["ciudades"][i]: dataframe["poblacion"][i] for i in range(len(dataframe["ciudades"])) }

for key, value in cities_dictionnary.items():
    print(f"City {key} has a population of {value}")


```
