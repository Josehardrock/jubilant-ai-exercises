```python
#•	Escribe una función que introduciendo una cadena de tamaño mayor que 7 devuelva los tres caracteres de en medio. 
# (Si la cadena tiene 7 devuelve el 3,4,5). La salida tiene que ser como en el ejemplo.
# "JhonDipPeta" → "Dip"
# “Jasonay" → "son"

def three_chars(chain):
    if len(chain) > 7:
        return chain[3:6]
    else:
        return chain[3:6]

print(f"String with more than 7 {three_chars('JhonDipPeta')}")
print(f"String with exactly 7 {three_chars('Jasonay')}")
print(f"String with another test {three_chars('extravagante')}")

#	Escribe una función que introduciendo una lista de tamaño par la devuelve en dos variables partida por la mitad como en el ejemplo.
# [1,2,3,4,5,6] -> [1,2,3], [4,5,6]
#["hello", "world", "I'm", "Lucas"] -> ["hello", "world"], ["I'm", "Lucas"]



def split_list(arglist):
    lis_size = len(arglist)

    list_index = lis_size // 2
    left_list = arglist[:list_index]
    right_list = arglist[list_index:]

    if len(arglist) % 2 != 0:
        print(f"Invalid list, list should be even")
    else:
        print(f"Strating list {arglist} : First part {left_list} ->  second par {arglist[list_index:]}") 

split_list([1,2,3,4,5,6])
split_list(["hello", "world", "I'm", "David"])
split_list(["hello", "world", "I'm", "Lucas", "I'm", "David", "Blabla"])
```
